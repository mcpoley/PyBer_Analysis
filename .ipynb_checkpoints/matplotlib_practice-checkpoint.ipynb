```python
 import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, color="Green", marker="*")
ax.plot(x_axis,y_axis, label='Boston')
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0, 45)
ax.set_xlabel('Dates')
ax.set_ylabel('Fares($)')
ax.grid()
ax.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis, label='Boston', color="green")
#Create labels for x and y axes 
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create a title
plt.title("Pybar Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis, y_axis)
```

```python
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
#Add a title
ax.set_title("Test")
#Add axes labels
ax.set_xlabel("Fares($)")
ax.set_ylabel("Dates")
plt.gca().invert_yaxis
ax.legend()
```

```python
plt.scatter(x_axis, y_axis, color="red", marker='x', s=y_axis)
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger, color="blue")
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis, color="skyblue",s=[i * 5 for i in y_axis], alpha=.2, edgecolors="black", marker='o')
ax.invert_xaxis()
```

```python
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
categories=['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
```

```python
plt.pie(revenue)
plt.show()
```

```python
plt.pie(revenue, labels=categories, autopct="%1.1f%%")
plt.show()
```

```python
explode=[0, 0, 0, 0, 0, .04]
```

```python
plt.pie(revenue, labels=categories, autopct="%1.1f%%", explode=explode)
plt.show()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, 
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred","tomato", "mistyrose"]
explode_values=(0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')
plt.show()
```
